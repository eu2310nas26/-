## 6. Поиск закономерностей, варианты решения бизнес-задачи

```python
genres_analysis = anime_cleaned.groupby("genres").agg(
    average_popularity=("popularity", "mean"),
    average_score=("score", "mean"),
    total_members=("members", "sum")
).reset_index()
genres_analysis = genres_analysis.sort_values(by="average_popularity", ascending=False)
genres_analysis
# обычно при посмотре чего-либо мы смотрим на средние оценки, выставленных от людей, которые уже посмотрели какое-либо аниме.
# с помощью total_members мы знаем сколько людей посмотрело данное аниме, average_score какая средняя оценка, average_popularity высчитывается средняя популярность.
# если жанр с высоким значением популярности также имеет высокyю среднюю оценку, это может указывать на то, что аниме также качественные.
# если высокие показатели популярности с низкими оценками, это может говорить о некачественности аниме.
```

```python
plt.figure(figsize = (12, 6))
sns.histplot(anime_cleaned["score"], bins=20, kde = True, color = "pink")
plt.title("распределение оценок аниме", fontsize = 16)
plt.xlabel("оценка", fontsize = 14)
plt.ylabel("частота", fontsize = 14)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.grid(axis = "y")
plt.show()
# показывает частоту просмотра аниме (тоесть аниме с высоким рейтингом видело мало, а с маленьким много)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 2)
genres_count = anime_cleaned["genres"].value_counts().head(10)  # топ-10 жанров
sns.barplot(x = genres_count.values, y = genres_count.index)
plt.title("топ-10 жанров аниме")
plt.xlabel("количество")
plt.ylabel("жанр")
plt.grid(axis = "x", linestyle = "--")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 2)
sns.regplot(data = anime_cleaned, x = "popularity", y = "score")
plt.title("зависимость оценки от популярности")
plt.xlabel("популярность")
plt.ylabel("оценка")
plt.grid()
plt.show()
# зависимость между популярностью и оценкой
```
